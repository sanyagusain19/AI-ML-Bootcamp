## Anime list 


</div>

```python
animes = [
     "Akira",
     "Spirited away",
     "Attack on titan",
      "Grave of the Fireflies",
      "Death note",
    "A Silent Voice",
    "Nausicaa of the Valley of the Wind",
    "Wolf Children",
    "My Neighbor Totoro",
    "The Girl Who Leapt Through Time" ,
    "fullmetal Alchemist" ,
    "Naruto",
    "One piece",
    "Violet evergarden",
    "steins gate",
    "Dr. stone"
]
api_key ="83306ba2"
```

<div style="background:#F7FCFF;padding:15px;border-radius:12px;">

##  Constructing the API Request and looping through anime list🌐

Creating a dynamic request URL by combining the anime title with the personal API key for authenticated access.

</div>

```python
anime_movie = []
for anime in animes:
   
    url = f"https://www.omdbapi.com/?t={anime}&apikey={api_key}"
    response = requests.get(url)
    data = response.json()
    if data['Response'] == 'True':
        anime_movie.append({
            "Title": data['Title'],
            "Year": data['Year'],
            "Genre": data['Genre'],
            "IMDb Rating": data['imdbRating'],
            "Director": data['Director'],
            'plot':data['Plot']
        })
    
    
```

# 🌸✨ Creating the DataFrame ✨🌸

Transforming the collected anime records into a structured **Pandas DataFrame**, making the dataset organized, searchable, and ready for analysis.

✨ From scattered API responses to one unified story.

---



*"Just as memories come together to tell a story, every row in this DataFrame becomes a chapter of the anime universe."* 🌸

</div>

```python
import pandas  as pd
df = pd.DataFrame(anime_movie)
df
```

```python
#as rating column is string nd have some n/a values therefore we can't use .astype(float), instead we use to_numeric which can handle error and then we can use fill na too...
df['IMDb Rating'] = pd.to_numeric(df['IMDb Rating'], errors='coerce')
df
```

```python
#fill na:
df['IMDb Rating'] = df['IMDb Rating'].fillna(0)
df
```

```python
#as some directors are unknown too:
df['Director'] = df['Director'].replace('N/A', 'unknown')
df
```
