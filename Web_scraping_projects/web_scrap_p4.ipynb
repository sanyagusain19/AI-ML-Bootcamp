<h2>📦 Import Libraries</h2>

<p>Loading the essential Python libraries required for web scraping, data manipulation, and visualization.</p>

</div>

```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
```

<div style="
background:#F9FAFB;
padding:15px;
border-left:8px solid #3B82F6;
border-radius:12px;">

<h2>🌐 Connect to Website</h2>

<p>Sending an HTTP request to retrieve the webpage for scraping.</p>

</div>

```python
headers = {"User-Agent": "Mozilla/5.0"}

url = f"https://myanimelist.net/topanime.php?limit=0"
response = requests.get(url, headers)
response.status_code
```

<div style="
background:#FFF7ED;
padding:15px;
border-left:8px solid #F59E0B;
border-radius:12px;">

<h2>🔍 Parse HTML</h2>

<p>Parsing the HTML content using BeautifulSoup to locate the required elements.</p>

</div>

```python
html_doc = response.text
```

```python
soup = BeautifulSoup(html_doc, 'html.parser')
print(soup.prettify())
```

<div style="
background:#F5F3FF;
padding:15px;
border-left:8px solid #9a79e7;
border-radius:12px;">

<h2> Extract Data</h2>

<p>Collecting titles, scores, rankings, URLs, and other relevant information from the webpage.</p>

</div>

```python
a =soup.find_all('tr', class_ = 'ranking-list')
a
rank = []
title =[]
score =[]
url = []
book_price = []
for row in a :
    
    rank.append( row.find("span", class_="top-anime-rank-text").text.strip())
    title.append(row.find("h3", class_="anime_ranking_h3").text.strip())
    score.append(row.find("td", class_="score").text.strip())
    url.append(row.find("a", class_ = "hoverinfo_trigger").get('href'))
    length = (row.find_all('span', class_ ="bookstore-price")) #find all returns a list thats whyyyy we can't use list
    if len(length) > 0:
        book_price.append(length[0].text.strip())
    else:
        book_price.append('N/A')

title
rank
url
book_price
```

```python
anime_detail ={
    'Title' : title,
    'Score': score,
    'Rank': rank,
    'Url':  url,
    'book_price': book_price
}
anime_detail
```

<div style="
background:#EFF6FF;

<h2>Build DataFrame 📊 </h2>

<p>Organizing the extracted information into a structured Pandas DataFrame.</p>

</div>

```python
df = pd.DataFrame(anime_detail)
df
```
