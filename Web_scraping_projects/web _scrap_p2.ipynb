<h1> <b>  Fetch url (Companies data) 🌐
</b></h1>
<h3> checking status code </h3>


</div>


```python
url = "https://www.ambitionbox.com/list-of-companies?page=1"
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'} 

response = requests.get(url, headers= headers)
response.status_code
```

<div style="background: #FAF5F7;
            padding: 10px;
            border-radius: 15px;
            color: pink;">


<h1> <b>  Storing html text inside a html_doc named variable
</b></h1>



</div>

```python
html_doc = response.text
```

<div style="background: #FAF5F7;
            padding: 10px;
            border-radius: 15px;
            color: pink;">


<h1> <b> Parse HTML using BeautifulSoup 
</b></h1>



</div>

```python
soup = BeautifulSoup(html_doc, 'html.parser')
print(soup.prettify())
```

<div style="background: #FAF5F7;
            padding: 10px;
            border-radius: 15px;
            color: pink;">


<h1> <b> Finding what's inside 'h' tag 
</b></h1>



</div>

```python
print(soup.find_all('h1')[0].text)
```

<div style="background: #FAF5F7;
            padding: 10px;
            border-radius: 15px;
            color: pink;">


<h1> <b> Finding what's inside 'h2' 
</b></h1>



</div>

```python
companies = soup.find_all('h2', class_="companyCardWrapper__companyName")
for  i in companies:
    print(i.text.strip())
    
```

```python
company = soup.find_all('div', class_= 'companyCardWrapper')
len(company)
```

<div style="background: #FAF5F7;
            padding: 10px;
            border-radius: 15px;
            color: pink;">


<h1> <b> Extracting Information about each company
</b></h1>



</div>

```python
name = []
Ratings=[]
locations=[]
worst_rated_for = []
best_rated_for = []
for i in company:
    name.append(i.find('h2').get_text(strip = True))
    Ratings.append(i.find('span',class_='companyCardWrapper__companyRatingCount').get_text(strip = True))
    locations.append(i.find_all('span')[2].get_text(strip = True))
    rating_value = i.find_all('span', class_='companyCardWrapper__ratingValues')
    worst_rated_for.append(rating_value[1].get_text(strip = True) if len(rating_value)> 1 else 'NA')
    best_rated_for.append(rating_value[1].get_text(strip = True) if len(rating_value)> 1 else ' NA')
    
locations
worst_rated_for
rating_value
final={
    'name': name,
    'rating':Ratings,
    'location':locations,
    'best_rated_for': best_rated_for,
    'worst_for':worst_rated_for
}

final
```

<div style="background: #FAF5F7;
            padding: 10px;
            border-radius: 15px;
            color: pink;">


<h1> <b> Preparing a DataFrame using pandas library
</b></h1>



</div>

```python
import pandas as pd
df = pd.DataFrame(final)
df.head()
```
