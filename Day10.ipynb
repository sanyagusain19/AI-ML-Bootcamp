# SQL[Structured Query Language]

## Why SQLite?
Because:
  - Lightweight
  - Serverless-> doesnot require a separate server installed.
- Check if python is installed? 
- Use built in library -> SQLite[Pre installed witth python > 2.5]


```python
import sqlite3 as sql
```

```python
# Connect with database[Create if doesnot exist]
connection = sql.connect('day10.db')
```

```python
# Cursor that traverse the records, create cursor object using connection
cursor = connection.cursor()
```

```python
cursor.execute('''CREATE TABLE students(
    id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    age INTEGER,
    city Text
);''')
```

```python
connection.commit()
```

```python
cursor.execute("""INSERT INTO students(name, age, city) VALUES ( 'Sana',20,'Yamuna nagar' ); """)
cursor.execute("""INSERT INTO students(name, age, city) VALUES('Sam',20,'Sonipat');""")
```

```python
connection.commit()
```

```python
cursor.execute('''SELECT * FROM students''')
```

```python
result = cursor.fetchall()
```

```python
for records in result:
    print(records)
```

```python
import pandas as pd
pd.DataFrame(result)
```

```python
import numpy as np
# from faker import faker
# faker = faker()
np.random.seed(30)

s= {
    # 'name' : [faker.name() for x in range (10000)], #site: https://towardsdatascience.com/build-your-own-synthetic-data-15d91389a37b/
    'name': np.random.choice(['Sneha','Murli','Misa','Nandini','Shweta'],10000),
    'age' : np.clip(np.random.normal(30,20,10000),19,30).round(1),
    'city':np.random.choice(['YNR','Sonipat','Panipat','Roorkie','Karnal'],10000)
} 
s

s_data = pd.DataFrame(s)
s_data
s_data.to_csv(" synthetic student small data", index = False)
```

```python
connection = sql.connect('synthetic.db')
```

```python
cursor = connection.cursor()
```

```python
cursor.execute('''CREATE TABLE student(
    
    name TEXT NOT NULL,
    age INTEGER,
    city Text
);''')
```

```python
connection.commit()
```

```python
a = list(s_data.itertuples(index= False, name = None))
```

```python
cursor.executemany("""INSERT INTO students(name, age, city) VALUES (?,?,? ); """, a)
```

```python
connection.commit()
```
