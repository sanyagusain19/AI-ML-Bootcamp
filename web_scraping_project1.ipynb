<h1> <b>  Importing required libraries 🌐 
</b></h1>



</div>

```python
from bs4 import BeautifulSoup
import pandas as pd
```

<div style="background: #FAF5F7;
            padding: 10px;
            border-radius: 15px;
            color: pink;">


<h1> <b> Parse HTML using BeautifulSoup 
</b></h1>



</div>

```python
soup = BeautifulSoup(html_doc, "html.parser")
print(soup.prettify())
```

<div style="background: #FAF5F7;
            padding: 10px;
            border-radius: 15px;
            color: pink;">


<h1> <b> Finding games information container
</b></h1>



</div>

```python
#scraping game name:
games = soup.find_all('a', class_="search_result_row")
games
```

<div style="background: #FAF5F7;
            padding: 10px;
            border-radius: 15px;
            color: pink;">


<h1> <b>  Extracting games name, price, release date and image url  
</b></h1>



</div>

```python
games_data=[]
from urllib.parse import urljoin
image_url =[]
for game in games:
   game_name = game.find('span',class_='title').get_text(strip = True)
   price_element = game.find("div", class_="discount_final_price").get_text(strip = True)
   release_element = game.find("div", class_="search_released").get_text(strip = True)
   
 
   image = game.find('img')
   src = image.get('src')
   
   games_data.append({'game_name': game_name,
      'price': price_element,
      'release': release_element,
      'image_url': urljoin(url,src)
   })
games_data
    
```

<div style="background: #FAF5F7;
            padding: 10px;
            border-radius: 15px;
            color: pink;">


<h1> <b>  Converting raw data to DataFrame 
</b></h1>



</div>

```python
df = pd.DataFrame(games_data)
```
