<h1 style="color:#2E8B57; font-size:42px;">
Importing necessary Libraries:
</h1>

</div>

```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
```

<div style="background:#EEF9F1; padding:22px; border-radius:18px;">

<h1 style="color:#2E8B57; font-size:42px;">
🎵 Fetching URL 
</h1>

</div>

```python
url = "https://open.spotify.com/playlist/37i9dQZEVXbMDoHDwVN2tF"
response = requests.get(url)
```

<div style="background:#EEF9F1; padding:22px; border-radius:18px;">

<h1 style="color:#2E8B57; font-size:42px;">
🎵 Storing response in a variable
</h1>

</div>

```python
print(response.status_code)
html_doc = response.text
```

<div style="background:#EEF9F1; padding:22px; border-radius:18px;">

<h1 style="color:#2E8B57; font-size:42px;">
Parsing HTML using BeautifulSoup:
</h1>

</div>

```python
soup = BeautifulSoup(html_doc, 'html.parser')
print(soup.prettify())
```

```python
info = soup.find_all('div',  class_="IJBqrvvMtwQ258xS")
```

<div style="background:#EEF9F1; padding:22px; border-radius:18px;">

<h1 style="color:#2E8B57; font-size:42px;">
Extracting Artist Names & Profile Links 🎵 
</h1>

</div>

```python
name = []
artist_link = []
for i in info:
    artist_link.append(i.find('a',class_ = "izvk_EbfhLUCNE6Q").get('href'))
    name.append(i.find('a', class_ = "izvk_EbfhLUCNE6Q").text.strip())
name 
artist_link
```

<div style="background:#EEF9F1; padding:22px; border-radius:18px;">

<h1 style="color:#2E8B57; font-size:42px;">
🎵 fetched artist details
</h1>

</div>


```python
artist_detail = {
    'artist_name': name,
    'artist_link':artist_link
}
artist_detail
```

<div style="background:#EEF9F1; padding:22px; border-radius:18px;">

<h1 style="color:#2E8B57; font-size:42px;">
Converting to dataframe 🎵 
</h1>

</div>

```python
df = pd.DataFrame(artist_detail)
df
```
