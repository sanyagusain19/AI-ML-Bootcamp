```python
import numpy as np
import pandas as pd
np.random.seed(45)

n_student =10000
data={
    "student_id" :[f"STU{i:05d}" for i  in range(1,n_student +1)] ,# max 5 decimal ...
    "gender":np.random.choice(["Male","Female"],n_student),
    "Age": np.random.randint(18,25,n_student),
    "Dept":np.random.choice(["CSE","IT","AIML","Mechanical"],n_student),
    "Residence":np.random.choice(["day_scholar","hostel"],n_student,p=[0.4,0.6]),
    "family_income":np.random.randint(50000,1000000,n_student),
    "attendance_percentage": np.clip(np.random.normal(75, 15, n_student), 20, 100).round(1),
    "sem_cgpa":np.clip(np.random.uniform(3.0,9.5,n_student),2,10).round(2),
    "internal_assesment":np.random.randint(85,100,n_student),
    # "backlogs":np.random.choice(["yes","No"],n_student,p=[0.6,0.4]),
    "backlogs":np.random.poisson(0.5,n_student),
    # "assignment_submission":np.random.choice(["yes","No"],n_student),
    "assignment_submission_rate": np.clip(np.random.normal(85,15,n_student),0,100).round(1),
    # "previous_exam_performance": np.random.normal(90,10,n_student),
    "lMS_login_freq":np.random.poisson(0.5,n_student),
    "Library_visit":np.random.poisson(5,n_student),
    "part_of_club":np.random.choice([0,1],n_student,p=[0.5,0.5]),
    "scholorship_status":np.random.choice([0,1],n_student,p=[0.8,0.2]),
    "cons_absent_days": np.random.randint(0,11,n_student),
    "Academic_warnings":np.random.choice([0,1,2,3],n_student,p=[0.7,0.1,0.1,0.1]),
    "Ttn_fee_due":np.random.choice([0,1],n_student,p=[0.2,0.8])
    }
```

```python
df=pd.DataFrame(data)
df
```

```python
#Risks score based on domain knowledge:
#calculate weighted risk score->
risk_score= (

    0.30 * (df["attendance_percentage"] < 60).astype(int) +  #0.30 is the contribution.
    0.25 * (df["sem_cgpa"]<5.5).astype(int) +
    0.15 * (df["Ttn_fee_due"]) +
    0.10 * (df["Academic_warnings"] >0).astype(int) +
    0.10 * (df["cons_absent_days"] >10).astype(int) +
    0.20 * (df["backlogs"] >+5).astype(int) -
    0.05 * (df["part_of_club"] ) -
    0.05 * (df["lMS_login_freq"] > 25).astype(int)

)
risk_score += np.random.normal(0,0.05,n_student)
df["drop_out"] = (risk_score >= 0.50).astype(int)
df["drop_out"]
```

```python
df.to_csv("student_dropout_synthetic_dataset" , index=False)
```
